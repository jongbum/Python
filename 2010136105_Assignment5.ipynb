```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return '[Count (step: '+str(self.step)+')]' + str(self.count)
        
    def incr(self):
        self.count += self.step
        
    def __call__(self):
        self.count += self.step
        return '[Count (step: '+str(self.step)+')]' + str(self.count)
        
    def __add__(self, x):
        self.count +=x
        return '[Count (step: '+str(self.step)+')]' + str(self.count)
        
    def __sub__(self, x):
        self.count -=x
        return '[Count (step: '+str(self.step)+')]' + str(self.count)
        
    def __cmp__(self, other):
        return self.count - other

        
c=Counter(10)
d=Counter(10,2)
print c
print d
print

c.incr()
d.incr()
print c
print d
print

c()
d()
print c
print d
print

print c+5
print d-5

print c>10
print d>10
print c<10
print d<10
print c==17
print d!=9
```
